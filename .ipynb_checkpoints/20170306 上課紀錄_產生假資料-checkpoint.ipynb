```python
plt.scatter(x,y,color="#8DFFCD",lw=3)
plt.plot(x,y)
```

```python
def f(x, a, b):
    return a*x + b
```

```python
from scipy.optimize import curve_fit
```

```python
popt, pcov = curve_fit(f, x, y) #p參數+opt最佳化 #p參數+cov
```
