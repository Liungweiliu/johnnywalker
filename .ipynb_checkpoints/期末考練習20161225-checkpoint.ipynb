```python
from ipywidgets.widgets import *
from IPython.display import display
```

```python
html = HTML('<h1>Hi</h1')
```

```python
display(html)
```

```python
html = HTML('<h1>GGininder</h1')
```

```python
display(html)
```

```python
from time import sleep
```

```python
worm = HTML('<div style="color:green;">oooo</div>')
```

```python
display(worm)
for i in range(80):
    htmlbase = '<div style="color:green;">%s</div>'
    html = htmlbase % ("&nbsp;" * (i+1) + "oooo")
    worm.value = html
    sleep(0.2)
```

```python
p = IntProgress()
```

```python
display(p)
```

```python
p.value=33
```

```python
display(p)

for i in range(101):
    p.value = i
    sleep(0.1)
```

```python
display(img_html)
```
