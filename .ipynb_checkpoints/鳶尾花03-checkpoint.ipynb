```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
y = iris.target
```

```python
x1 = iris.data[:, :1]
x2 = iris.data[:, 1:2]
x3 = iris.data[:, 2:3]
x4 = iris.data[:, 3:4]
x12 = x1-x2
x34 = x3-x4
```

```python
x12d34 = np.c_[x12,x34]
```

```python
clf.fit(x12d34, y)
```

```python
plt.scatter(x12,x34, s=50, c=clf.predict(x12d34),cmap=plt.cm.Wistia)
```
