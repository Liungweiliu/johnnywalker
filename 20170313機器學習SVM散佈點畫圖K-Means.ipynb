```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[-2,3], [-8,7], [1,-2], [2,-2]])
y = np.array([1,1,2,2])
```

```python
plt.plot(x,y)
```

```python
x[:, 0]
```

```python
x[:, 1]
```

```python
plt.scatter(x[:, 0], x[:, 1], s=50, c=y)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x, y)
```

```python
clf.predict([[-2,3]])
```

```python
clf.predict([[1,2]])
```

```python
clf.predict([[1,-3]])
```

```python
#x=-10~4 y=-4~8
```

```python
gd = np.array([(i, j) for i in np.arange(-10, 4,0.6)
              for j in np.arange(-4, 8,0.6)])
```

```python
gdc = clf.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc)
```

```python
#K-Means電腦自分法
```

```python
x = np.random.rand(100, 2)
```

```python
plt.scatter(x[:, 0], x[:, 1], s=50, color="#87FF65")
```

```python
p = np.random.rand(2500, 2)
plt.scatter(p[:, 0], p[:, 1], s=5, color="#4C4CE0")
```

```python
from sklearn.cluster import KMeans
```

```python
#開一台機器 分7群
clf = KMeans(n_clusters=7)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:, 0], x[:, 1], s=50, c=clf.labels_)
```

```python
clf.predict([[1000000,-30000000]])
```

```python
gd = np.array([(i, j) for i in np.arange(0, 1,0.1)
              for j in np.arange(0, 1,0.1)])
```

```python
plt.scatter(gd[:, 0], gd[:, 1], s=50, c=clf.predict(gd))
```
