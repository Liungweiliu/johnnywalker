```python
import pandas_datareader.data as web #抓股市資料
import datetime
```

```python
symbol = "AAPL"#蘋果
day_start = datetime.datetime(2010,1,1)#開始日
day_end = datetime.datetime(2016,11,28)#結束日
```

```python
df4 = web.DataReader(symbol, 'yahoo', day_start, day_end)
```

```python
df4.head()
```

```python
df4.head
```

```python
df4.Close.plot()#收盤價走勢圖
```

```python
#取差價 後項減前項
```

```python
df4["Diff"] = df4.Close.diff()
```

```python
df4.head()
```

```python
df4.Diff.plot()
```

```python
#後項減前項/前項
```

```python
df4["Rel_Diff"] = df4.Diff / df4.Close
```

```python
df4.head()
```

```python
df4.Rel_Diff.plot()
```

```python
df4.dropna(inplace=True)##inplace=True 回存 dropna除去NA
```

```python
df4
```

```python
df4.head()
```

```python
df4.Diff.min()
```

```python
#刪特殊資料
```

```python
df4[df4.Diff == df4.Diff.min()]
```

```python
df4.Diff.argmin()#arg抓座標 argmin抓最小值 argmax抓最大值
```

```python
df4.drop(df4.Diff.argmin(),inplace=True) #drop刪資料 inplace回存
```
