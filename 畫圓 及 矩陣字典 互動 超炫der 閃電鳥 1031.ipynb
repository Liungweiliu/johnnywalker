```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
r = 3
t = np.linspace(-2*np.pi, 2*np.pi, 200)

x = r*np.cos(t)
y = r*np.sin(t)

plt.plot(x, y, lw=3,c='#F0F987')
ax = plt.gca()
ax.set_aspect('equal')
```
